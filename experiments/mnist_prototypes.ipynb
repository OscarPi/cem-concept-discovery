```python
import numpy as np
from cemcd.data import mnist
import matplotlib.pyplot as plt
```

```python
def plot_image_grid(images):
    n_images = len(images)
    if n_images != 16:
        raise ValueError("Expected 16 images, but got {}".format(n_images))

    grid_size = 4
    _, axes = plt.subplots(grid_size, grid_size, figsize=(8, 8))
    axes = axes.flatten()

    for i, img in enumerate(images):
        img = np.transpose(img.cpu().detach().numpy(), (1, 2, 0))
        axes[i].imshow(img)
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
mnist_datasets = mnist.MNISTDatasets(n_digits=2, max_digit=6)
data_getter = mnist_datasets.train_getter
```

```python
discovered_concept_labels = np.load(
    "/cem-concept-discovery/results/mnist_add-ALPHA-ALPHA-PAPA/discovered_concepts.npz"
)["discovered_concept_labels"][:, 87]
```

```python
images = []
while len(images) < 16:
    i = np.random.randint(0, discovered_concept_labels.shape[0])
    if discovered_concept_labels[i] == 0:
        continue
    images.append(data_getter(i)[0])
```

```python
plot_image_grid(images)
```
